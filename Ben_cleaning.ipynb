## Fuel Df Cleaning

```python
fuel_cols = ['month', 'airline_id', 'unique_carrier', 'carrier', 'carrier_name', 'carrier_group_new', 'sdomt_gallons', 'satl_gallons', 'spac_gallons', 'slat_gallons', 'sint_gallons', 'ts_gallons', 'tdomt_gallons', 'tint_gallons', 'total_gallons', 'sdomt_cost', 'satl_cost', 'spac_cost', 'slat_cost', 'sint_cost', 'ts_cost', 'tdomt_cost', 'tint_cost', 'total_cost', 'year']
fuel_df = pd.read_csv('data/fuel_consumption.csv', header=None, names=fuel_cols)
```

```python
fuel_df.head()
```

```python
fuel_df.info()
```

```python
# check for duplicates
fuel_df.duplicated().sum()
```

### Address null values

```python
# check null values
fuel_df.isna().sum()
```

```python
fuel_df[fuel_df['carrier']=='0JQ']
```

```python
# First row missing values can be matched to row beneath it
fuel_df.iloc[0,2] = '0JQ'
fuel_df.iloc[0,1] = 21236
fuel_df.iloc[0,4] = 'Vision Airlines'
```

```python
fuel_df[fuel_df['tint_cost'].isna()]
```

```python
fuel_df['tint_cost'].fillna(value=(fuel_df['total_cost']-fuel_df['tdomt_cost']), inplace=True)
```

```python
fuel_df[fuel_df['total_gallons'].isna()][['sdomt_gallons', 'satl_gallons', 'spac_gallons', 'slat_gallons', 'sint_gallons', 'ts_gallons', 'tdomt_gallons', 'tint_gallons', 'total_gallons']]
```

```python
fuel_df['total_gallons'].fillna(value=(fuel_df['tdomt_gallons'] + fuel_df['tint_gallons']), inplace=True)
```

```python
fuel_df[fuel_df['tdomt_gallons'].isna()][['sdomt_gallons', 'satl_gallons', 'spac_gallons', 'slat_gallons', 'sint_gallons', 'ts_gallons', 'tdomt_gallons', 'tint_gallons', 'total_gallons']]
```

```python
fuel_df['tdomt_gallons'].fillna(value=(fuel_df['total_gallons'] - fuel_df['tint_gallons']), inplace=True)
```

```python
fuel_df.shape
```

```python
no_nulls = fuel_df.dropna()
no_nulls.shape
```

```python
fuel_df=no_nulls
```
